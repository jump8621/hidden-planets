# Read the CSV and Perform Basic Data Cleaning

```python
df = pd.read_csv("exoplanet_data.csv")
# Drop the null columns where all values are null
df = df.dropna(axis='columns', how='all')
# Drop the null rows
df = df.dropna()
df.head()
```

# Select your features (columns)

```python
# Set features. This will also be used as your x values.
#selected_features = df[[]]

selected_features = df[['koi_tce_plnt_num', 'koi_fpflag_nt', 'koi_fpflag_ss', 'koi_fpflag_co', 'koi_fpflag_ec', 'ra', 'dec', 'koi_srad', 'koi_srad_err1', 'koi_srad_err2', 'koi_steff', 'koi_steff_err1', 'koi_steff_err2', 'koi_model_snr', 'koi_slogg', 'koi_slogg_err1', 'koi_slogg_err2', 'koi_kepmag', 'koi_period', 'koi_period_err1', 'koi_period_err2', 'koi_time0bk', 'koi_time0bk_err1', 'koi_time0bk_err2', 'koi_depth', 'koi_depth_err1', 'koi_depth_err2', 'koi_prad', 'koi_prad_err1', 'koi_prad_err2', 'koi_teq', 'koi_insol', 'koi_insol_err1', 'koi_insol_err2', 'koi_duration', 'koi_duration_err1', 'koi_duration_err2', 'koi_impact', 'koi_impact_err1', 'koi_impact_err2']]
#,,
```

# Create a Train Test Split

Use `koi_disposition` for the y values

```python
#possibly the stuff below - but not sure - check on koi_disposition 
#A Kepler object of interest (KOI) is a star observed by the Kepler space telescope
# that is suspected of hosting one or more transiting planets.

# X = brain[["gender", "age", "size"]]
# y = brain["weight"].values.reshape(-1, 1)
# print(X.shape, y.shape)

# X = lsd.tc.values.reshape(-1, 1)
# y = lsd.score.values.reshape(-1, 1)
# print("Shape: ", X.shape, y.shape)
# X  
X = selected_features
y = df['koi_disposition']
print(X.shape, y.shape)

# data = X.copy().values.reshape(-1,1)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1, stratify=y)
```

```python
X_train.head()
```

# Pre-processing

Scale the data using the MinMaxScaler and perform some feature selection

```python
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
import tensorflow
from tensorflow.keras.utils import to_categorical
from sklearn.model_selection import train_test_split
```

```python

X_scaler = MinMaxScaler().fit(X_train)

X_train_scaler = X_scaler.transform(X_train)
X_test_scaler = X_scaler.transform(X_test)
```

```python
# Step 1: Label-encode data set
label_encoder = LabelEncoder()
label_encoder.fit(y_train)
encoded_y_train = label_encoder.transform(y_train)
encoded_y_test = label_encoder.transform(y_test)

# Step 2: Convert encoded labels to one-hot-encoding
y_train_categorical = to_categorical(encoded_y_train)
y_test_categorical = to_categorical(encoded_y_test)

y_test_categorical

```

```python
y_train_categorical.shape
```

# Train the Model



```python
# first, create a normal neural network with 2 inputs, 6 hidden nodes, and 2 outputs

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# model = Sequential()
# model.add(Dense(units=6, activation='relu', input_dim=40))
# #prediction? the output = has to be 2 neurons/units because 2 classes wanting to predict
# model.add(Dense(units=3, activation='softmax'))


model = Sequential()
model.add(Dense(units=6, activation='relu', input_dim=40))
model.add(Dense(units=6, activation='relu'))
model.add(Dense(units=3, activation='softmax'))
```

```python
model.summary()
```

```python
# Compile the model

model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
# Fit the model to the training data

model.fit(
    X_train_scaler,
    y_train_categorical,
    epochs=100,
    shuffle=True,
    verbose=2
)
```

```python
model_loss, model_accuracy = model.evaluate(
    X_test_scaler, y_test_categorical, verbose=2)
print(
    f"Normal Neural Network - Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
encoded_predictions = model.predict_classes(X_test_scaler[:5])
prediction_labels = label_encoder.inverse_transform(encoded_predictions)
```

```python
print(f"Predicted classes: {prediction_labels}")
print(f"Actual Labels: {list(y_test[:5])}")
```
